# Conditional Filtering

```python
df['Home.Ownership'].value_counts()
```

```python
df['Home.Ownership']== 'MORTGAGE'
```

```python
# get all the records where home ownership is mortgage

df[df['Home.Ownership'] == 'MORTGAGE']
```

```python
# get all the records where the state is tx
df[df['State']=='TX']
```

```python
# get all texas records  where home ownership is mortgage

df[( df['State']=='TX') & (df['Home.Ownership'] == 'MORTGAGE')].shape
```

```python
# get all the records where state is texas or new jersy

df[(df['State']=='TX') | (df['State']=='NJ')].shape
```

```python
df[df['State']. isin (['NJ','TX'])].shape       # we can use .isin also insted of | this
```

```python
# get all the records where home owership is not mortage
df[~(df['Home.Ownership']=='MORTGAGE')].shape
```

```python
# get all the records where state is NJ SC or NY

df[df['State'].isin(['NJ','SC' ]) |(['NY'])].shape
```

```python
# Repeat above with np.where

df.iloc[np.where(df['State'].isin(['NJ','SC']) | (['NY']))].shape
```

```python
# GEt NJ records with loan purpose as either home_improvement/major purchase and monthly income<5000

df[(df['State']=='NJ') & (df['Loan.Purpose'].isin(['major_purchase','home_improvement']))&(df['Monthly.Income']<5000)]
```

```python
# GEt non NJ records with loan purpose as either home_improvement/major purchase and monthly income<5000
df[~(df['State']=='NJ') & ~(df['Loan.Purpose'].isin(['major_purchase','hoome_improvement']))&(df['Monthly.Income']<5000)]
```

```python
# GEt NJ records with loan purpose as either home_improvement/major purchase and monthly income<1000
df[~(df['State']=='NJ') & ~(df['Loan.Purpose'].isin(['major_purchase','hoome_improvement']))&(df['Monthly.Income']<1000)].shape
```
